```python
import geopandas as gpd
import matplotlib.pyplot as plt
import folium
```

```python
gdf = gpd.read_file('data/Shapes_Ae/Manzana.shp')
gdf_mpos = gpd.read_file('data/Shapes_AreasProd/mun.shp')
```

```python
gdf = gpd.overlay(gdf, gdf_mpos, how='intersection', keep_geom_type=True)
gdf.to_crs(crs='EPSG:4326')
gdf.explore(column='AE', cmap='spring', tooltip=['NOM_MUN', 'CULT', 'AE', 'AE_HA'])
```

```python
def style_function_mpos(feature):
    return {
        'fillOpacity': 0,  # Make the fill transparent
        'color': 'black',  # Black border color
        'weight': 1.5  # Border thickness
    }
```

```python
cultivo = 'Manzana'
m = folium.Map(location=(24.148794,-104.8425257), zoom_start=8, tiles='cartodb positron')
#mpos = folium.GeoJson(gdf_mpos).add_to(m)
mpos = folium.GeoJson(gdf_mpos, name='Municipios', style_function=style_function_mpos).add_to(m)

tooltip = folium.features.GeoJsonTooltip(fields=['CULT', 'NOM_MUN', 'AE'],
                               aliases=['Cultivo:', 'Municipio', 'AE:'],
                               localize=True,
                               labels=True,
                               sticky=True,
                               style="""
                                   background-color: #F0EFEF;
                                   border: 1.5px solid black;
                                   border-radius: 5px;
                                   box-shadow: 5px;
                                   """,
                            )
gjson = folium.GeoJson(gdf, name=cultivo, tooltip=tooltip).add_to(m)
folium.LayerControl().add_to(m)

layer_control = folium.LayerControl(collapsed=False, title='Capas').add_to(m)
m.save('index.html')
```
